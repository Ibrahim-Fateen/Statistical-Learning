```python
import pandas as pd

data = pd.read_csv('../Data/Auto.csv')
data = data.set_index('name')
data.head()
```

```python
quantitative_predictors = data.drop(columns=['cylinders', 'origin'])
quantitative_predictors.head()
```

```python
qualitative_predictors = data[['cylinders', 'origin']]
qualitative_predictors.head()
```

```python
quantitative_predictors.describe()
```

```python
qualitative_predictors.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
quantitative_predictors.hist(bins=20, figsize=(14, 10), layout=(2, 3))
plt.suptitle('Histograms of quantitative predictors')
plt.show()
```

```python
sns.pairplot(quantitative_predictors)
plt.suptitle('Pairplot of quantitative predictors', y=1.02)
plt.show()
```

```python
qualitative_predictors = ['cylinders', 'origin']
for predictor in qualitative_predictors:
    sns.boxplot(x=predictor, y='mpg', data=data)
    plt.title(f'Boxplot of {predictor} vs mpg')
    plt.show()
```
